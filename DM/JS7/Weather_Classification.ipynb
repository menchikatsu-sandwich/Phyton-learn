```python
import numpy as np
import pandas as pd
from sklearn import naive_bayes
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import ComplementNB
from sklearn.naive_bayes import CategoricalNB
```

```python
#Data Collection
#mengimpor dataset dan menyimpan dalam dataframe
path = "..\Dataset\API_Kaggle\weather_classification_data.csv"
df = pd.read_csv(path)
df
```

```python
#Data Preprocessing
#mengubah nama kolom
new_names={'Precipitation (%)': 'Precipitation','Visibility (km)': 'Visibility'}
df.rename(columns=new_names, inplace=True)
#perintah inplace = True akan menyimpan perubahan yang dilakukan
#sehingga jika ditampilkan, dataframe tetap akan menyimpan perubahan tersebut
df
```

```python
# cek NaN values
df.isna().sum()
```

```python
# cek tipe data
df.info()
```

```python
#Data Transformation
#mengubah kolom dengan tipe data object
#menjadi data numerik
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
feature = ['Cloud Cover', 'Season', 'Location' ]
df[feature] = df[feature].apply(le.fit_transform)
df
```

```python
from sklearn.preprocessing import MinMaxScaler
#Memilih kolom fitur yang dinormalisasi
X = df.iloc[:,:10]

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(X)
df_scaled
```

```python
# memilh kolom target
y = df['Weather Type']
y.head()
```

```python
#Membagi dataset menjadi data latih dan data uji
#dengan perbandingan 70% : 30%
from sklearn.model_selection import train_test_split

X_trainset, X_testset, y_trainset, y_testset = train_test_split(df_scaled, y, test_size=0.3, random_state=3)

print ("Ukuran X_trainset = ", X_trainset.shape)
print ("Ukuran y_trainset = ", y_trainset. shape)

print ("Ukuran X_testset = ", X_testset.shape)
print ("Ukuran y_testset = ", y_testset.shape)
```

```python
#Data Classification
#Membangun model klasifikasi dengan Bernoulli Naïve Bayes
WeatherTree = BernoulliNB(alpha=1.0, force_alpha=True, binarize=0.0,
fit_prior=True, class_prior=None)

#Melatih model dengan data latih
WeatherTree.fit(X_trainset,y_trainset)

#Memprediksi data uji dengan model
y_predWeather = WeatherTree.predict(X_testset)

#Menampilkan Label uji yang seharusnya
#dan label hasil prediksi
result = pd. DataFrame({'True Label': y_testset, 'Predicted Label': y_predWeather})
result.head()

#Menghitung nilai akurasi prediksi model
from sklearn import metrics
print("Bernoulli Naïve Bayes Accuracy: ", metrics.accuracy_score(y_testset, y_predWeather) )
```

```python
#Data Classification
#Membangun model klasifikasi dengan Gaussian Naïve Bayes
WeatherTree = GaussianNB(priors=None, var_smoothing=1e-09)

#Melatih model dengan data latih
WeatherTree.fit(X_trainset,y_trainset)

#Memprediksi data uji dengan model
y_predWeather = WeatherTree.predict(X_testset)

#Menampilkan label uji yang seharusnya
#dan Label hasil prediksi
result = pd.DataFrame({'True Label': y_testset, 'Predicted Label': y_predWeather})
result.head()

#Menghitung nilai akurasi prediksi model
from sklearn import metrics
print("Gaussian Naïve Bayes: ", metrics.accuracy_score(y_testset, y_predWeather))
```

```python
#Data Classification
#Membangun model klasifikasi dengan Multinomial Naive Bayes
WeatherTree = MultinomialNB(alpha=1.0, force_alpha=True, fit_prior=True, class_prior=None)

#Melatih model dengan data latih
WeatherTree.fit(X_trainset,y_trainset)

#Memprediksi data uji dengan model
y_predWeather = WeatherTree.predict(X_testset)

#Menampilkan Label uji yang seharusnya
#dan label hasil prediksi
result = pd. DataFrame({'True Label': y_testset, 'Predicted Label': y_predWeather})
result.head()

#Menghitung nilai akurasi prediksi model
from sklearn import metrics
print("Multinomial Naïve Bayes: ", metrics. accuracy_score(y_testset, y_predWeather))
```

```python
#Data Classification
#Membangun model klasifikasi dengan Complement Naïve Bayes
WeatherTree = ComplementNB(alpha=1.0, force_alpha=True, fit_prior=True, class_prior=None, norm=False)

#Melatih model dengan data latih
WeatherTree.fit(X_trainset,y_trainset)

#Memprediksi data uji dengan model
y_predWeather = WeatherTree.predict(X_testset)

#Menampilkan Label uji yang seharusnya
#dan label hasil prediksi
result = pd. DataFrame({'True Label': y_testset, 'Predicted Label': y_predWeather})
result.head()

#Menghitung nilai akurasi prediksi model
from sklearn import metrics
print("Complement Naïve Bayes: ", metrics.accuracy_score(y_testset, y_predWeather) )
```

```python
#Data Classification
#Membangun model klasifikasi dengan Categorical Naïve Bayes
WeatherTree = CategoricalNB(alpha=1.0, force_alpha=True, fit_prior=True,
class_prior=None, min_categories=None)

#Melatih model dengan data latih
WeatherTree.fit(X_trainset,y_trainset)

#Memprediksi data uji dengan model
y_predWeather = WeatherTree.predict(X_testset)

#Menampilkan Label uji yang seharusnya
#dan label hasil prediksi
result = pd. DataFrame({'True Label': y_testset, 'Predicted Label': y_predWeather})
result.head()

#Menghitung nilai akurasi prediksi model
from sklearn import metrics
print("Categorical Naïve Bayes: ", metrics.accuracy_score(y_testset, y_predWeather))
```
