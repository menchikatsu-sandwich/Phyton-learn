```python
df_grup = df.groupby('Weather Type')[["Humidity"]].mean()
df_grup
```

```python
import matplotlib.pyplot as plt

df_pie = df_grup.reset_index()
plt.figure(figsize=(6,6))

plt.pie(
    df_pie['Humidity'], 
    labels=df_pie['Weather Type'], 
    autopct='%1.2f%%', 
    explode=(0,0.1,0,0))
plt.title('Mean Humidity berdasarkan Weather Type', 
          pad=30, 
          loc='center', 
          fontsize=15, 
          color='blue')
plt.show()
```

```python
import numpy as np
df_bar = df.groupby("Weather Type")[['Humidity']].mean().reset_index()
x_coords = np.arange(len(df_bar["Weather Type"]))

plt.figure(figsize=(6,4))
plt.bar(x_coords, df_bar['Humidity'], tick_label=df_bar['Weather Type'])
plt.xticks(rotation=90)
plt.ylabel("Humidity")
plt.title("Mean Humidity berdasarkan Weather Type")
plt.show()
```

```python
df_bar_sort = pd.DataFrame({
    "Weather Type": df_bar["Weather Type"],
    "Humidity": df_bar["Humidity"]
})
df_bar_sort.sort_values(by="Humidity", inplace=True)
x_coords = np.arange(len(df_bar_sort["Weather Type"]))

plt.figure(figsize=(6,4))
plt.bar(x_coords, df_bar_sort['Humidity'], tick_label=df_bar_sort['Weather Type'])
plt.xticks(rotation=90)
plt.ylabel("Humidity")
plt.title("Mean Humidity berdasarkan Weather Type")
plt.show()
```
