```python
import pandas as pd
path = "..\Dataset\API_Kaggle\weather_classification_data.csv"
df = pd.read_csv(path)
df
```

```python
new_names = {'Precitipation (%)': 'Precipitation',
             'Visibility (km)': 'Visibility',}
df.rename(columns=new_names, inplace=True)
# perintah inplace = True akan menyimpan perubahan pada dataframe df
df
```

```python
# perintah isna() untuk mengecek NaN pada df
# perintah sum() untuk menjumlahkan jumlah NaN
df.isna().sum()
```

```python
import numpy as np

# membuat dataframe df_change baru yang berisi data dari dataframe df
df_new=df.copy()

# mengubah nilai "Winter pada kolom Season dengan nilai NaN
df_new["Season"] = df_new["Season"].replace("Winter", np.NaN)

# mengubah nilai negatif pada kolom Temperature dengan nilai NaN
df_new. Temperature=df_new. Temperature.mask(df_new.Temperature.lt(0),np.NaN)

df_new
```

```python
# perintah isna() untuk mengecek NaN pada df
# perintah sum() untuk menjumlahkan jumlah NaN
df_new.isna().sum()
```

```python
# perintah dropna() menghapus sel yang kosong
df_drop_nan = df_new.dropna()
df_drop_nan
```

```python
# perintah isna() untuk mengecek NaN pada df
# perintah sum() untuk menjumlahkan jumlah NaN
df_drop_nan.isna().sum()
```

```python
#perintah mode() menentukan nilai modus dari df
#penambahan [0] karena mode() mengembalikan nilai series
#walaupun hanya satu value
#sehingga perlu diset indeks ke-0 yang diambil
modus = df_new['Season'].mode() [0]
modus
```

```python
# perintah fillna mengisi nilai modus ke seluruh nilai kosong dari df
df_new["Season"] = df_new["Season"].fillna(modus)
df_new
```

```python
# perintah isna() untuk mengecek NaN pada df
# perintah sum() untuk menjumlahkan jumlah NaN
df_new.isna().sum()
```

```python
#perintah mean() untuk mencari rata-rata dari data
#perintah fillna untuk mengganti missing value dengan mean
df_mean = df_new.copy()
df_mean["Temperature"] = df_mean["Temperature"].fillna(value=df_mean["Temperature"].mean())
df_mean
```

```python
#Mengatasi nilai missing value dengan imputasi suka-suka
#Nilai suka-suka yang diberikan pada contoh adalah 99
df_arbitrary = df_new.copy()
df_arbitrary["Temperature"] = df_arbitrary["Temperature"].fillna(value=99)
df_arbitrary
```
