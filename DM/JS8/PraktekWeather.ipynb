```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
```

```python
path = "..\Dataset\API_Kaggle\weather_classification_data.csv"
df = pd.read_csv(path)
df
```

```python
new_names = {'Precipitation (%)': 'Precipitation',
             'Visibility (km)': 'Visibility',}
df.rename(columns=new_names, inplace=True)
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
feature = ['Cloud Cover', 'Season', 'Location']
df[feature] = df[feature].apply(le.fit_transform)
df
```

```python
from sklearn.preprocessing import MinMaxScaler
x = df.iloc[:, :10]

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(x)
df_scaled
```

```python
y = df['Weather Type']
y.head()
```

```python
#Membagi dataset menjadi data latih dan data uji
#dengan perbandingan 70% : 30%
from sklearn.model_selection import train_test_split

X_trainset, X_testset, y_trainset, y_testset = train_test_split(df_scaled, y, test_size=0.3, random_state=3)

print ("Ukuran X_trainset = ", X_trainset.shape)
print ("Ukuran y_trainset = ", y_trainset. shape)

print ("Ukuran X_testset = ", X_testset. shape)
print ("Ukuran y_testset = ", y_testset.shape)
```

```python
#Data Classification
#Membangun model klasifikasi dengan MLP Classifier
WeatherTree = MLPClassifier(hidden_layer_sizes=(100,), activation='relu',
solver='adam', alpha=0.0001, batch_size='auto',
learning_rate='constant', learning_rate_init=0.001,
power_t=0.5, max_iter=200, shuffle=True, random_state=None,
tol=0.0001, verbose=False, warm_start=False, momentum=0.9,
nesterovs_momentum=True, early_stopping=False,
validation_fraction=0.1, beta_1=0.9, beta_2=0.999,
epsilon=1e-08, n_iter_no_change=10, max_fun=15000)

#Melatih model dengan data latih
WeatherTree.fit(X_trainset,y_trainset)

#Memprediksi data uji dengan model
y_predWeather = WeatherTree.predict(X_testset)

#Menampilkan Label uji yang seharusnya
#dan Label hasil prediksi
result = pd. DataFrame({'True Label': y_testset, 'Predicted Label': y_predWeather})
result.head()

#Menghitung nilai akurasi prediksi model
from sklearn import metrics
print("MLP Classifier Accuracy: ", metrics.accuracy_score(y_testset, y_predWeather))
```

```python
#mengubah target label Y dari kategori
#menjadi biner
from tensorflow.keras.utils import to_categorical

encoder = LabelEncoder()
encoder.fit(y_trainset)
y_train = encoder.transform(y_trainset)
y_test = encoder.transform(y_testset)
y_train = to_categorical(y_train, num_classes=4)
y_test = to_categorical(y_test, num_classes=4)
print(y_train)
print(y_test)
```

```python
# membangun model
from keras.models import Sequential
from keras.layers import Flatten, Dense
from sklearn.metrics import confusion_matrix

model = Sequential()
model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dense(4,activation='softmax'))

model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['acc'])

# melatih model
model.fit(X_trainset,y_train,epochs=100,batch_size=10)

# mengevaluasi model
model. summary()

loss, accuracy = model.evaluate(X_testset, y_test)
print('ANN Loss: ', loss)
print('ANN Accuracy:', accuracy)
```
