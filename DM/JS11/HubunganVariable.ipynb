```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-6.0, 6.0, 0.1)

#Kemiringan dan intersep dapat disesuaikan untuk
#memverifikasi perubahan dalam grafik
y = 3*(x) + 2
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise

plt.plot(x, ydata, 'bo')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

```python
X = np.arange(1.0, 10000.0, 0.1)

# Persamaan y = ln(x) : log berbasis e
# Untuk log basis 10, dapat menggunakan np.log10(X)
Y = np. log (X)

plt.plot(X,Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable' )
plt.show()
```
