```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pylab as pl
%matplotlib inline
```

```python
path = "..\Dataset\API_Kaggle\Student_Performance.csv"
df = pd.read_csv(path)
df
```

```python
#Data Transformation
#mengubah kolom dengan tipe data object
#menjadi data numerik
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
feature = ['Extracurricular Activities']
df[feature] = df[feature].apply(le.fit_transform)
df
```

```python
X = df.iloc[:,:5]
y = df.iloc[:,5:]
```

```python
X.corr()
```

```python
#Membagi dataset menjadi data latih dan data uji
#dengan perbandingan 70% : 30%

from pandas. core. common import random_state
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

print ("Ukuran X_trainset = ", X_train.shape)
print ("Ukuran y_trainset = ", y_train. shape)

print ("Ukuran X_testset = ", X_test.shape)
print ("Ukuran y_testset = ", y_test.shape)
```

```python
#Data Classification
#Membangun model regresi linier sederhana
#https://scikit-learn.org/stabl|e/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(criterion='squared_error', splitter='best', max_depth=None,
min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0,
max_features=None, random_state=None, max_leaf_nodes=None,
min_impurity_decrease=0.0, ccp_alpha=0.0, monotonic_cst=None)

regressor. fit(X_train, y_train. values. ravel() )

#Memprediksi data uji dengan model
y_pred_test = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train)

hasil = {
"ActualValue": y_test.values. flatten(),
"PredictedValue": y_pred_test. flatten(),
}

#load data into a DataFrame object:
dfhasil = pd. DataFrame(hasil)
dfhasil
```

```python
#Model Evaluation
#mengevaluasi model regresi dengan metriks:
#MAE, MSE, RMSE dan R Squared Score dengan function

from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

8

mae = mean_absolute_error(y_test, y_pred_test)
mape = mean_absolute_percentage_error(y_test, y_pred_test)
mse = mean_squared_error(y_test, y_pred_test)
rmse = np. sqrt (mse)
r_squared = r2_score(y_test, y_pred_test)

print(f"Mean Absolute Error (MAE): {mae:,.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape*100:,.2f}%")
print(f"Mean Squared Error (MSE): {mse:,.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:,.2f}")
print(f"R squared score: {r_squared:.2f}")
```

```python
from sklearn.model_selection import GridSearchCV

# Inisialisasi model dasar dengan random_state = 42
regressor = DecisionTreeRegressor(random_state=42)

# Menentukan grid parameter (5 nilai untuk masing-masing parameter)
param_grid = {
'max_depth': [3, 5, 7, 9, None],
'min_samples_split': [2, 5, 10, 15, 20],
'min_samples_leaf': [1, 2, 4, 6, 8]
}

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(estimator=regressor,
param_grid=param_grid,
cv=5,
scoring='r2',
n_jobs =- 1,
verbose=1)              # 5-fold cross-validation
                        # atau bisa pakai 'neg_mean_squared_error'
                        # paralel semua core

# Melatih grid search dengan data latih
grid_search. fit(X_train, y_train. values. ravel() )

# Menampilkan parameter terbaik dan skor terbaik
print("Best Parameters:", grid_search.best_params_)
print("Best R2 Score:", grid_search.best_score_)

# Gunakan model terbaik untuk prediksi
best_model = grid_search.best_estimator_

y_pred_test = best_model.predict(X_test)
y_pred_train = best_model.predict(X_train)

# Buat DataFrame hasil prediksi
hasil = {
"ActualValue": y_test.values. flatten(),
"PredictedValue": y_pred_test.flatten(),
}

dfhasil = pd. DataFrame(hasil)
dfhasil
```

```python
#Model Evaluation
#mengevaluasi model regresi dengan metriks:
#MAE, MSE, RMSE dan R Squared Score dengan function

from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred_test)
mape = mean_absolute_percentage_error(y_test, y_pred_test)
mse = mean_squared_error(y_test, y_pred_test)
rmse = np.sqrt (mse)
r_squared = r2_score(y_test, y_pred_test)

print(f"Mean Absolute Error (MAE): {mae:,.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape*100:,.2f}%")
print(f"Mean Squared Error (MSE): {mse:,.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:,.2f}")
print(f"R squared score: {r_squared:.2f}")
```
