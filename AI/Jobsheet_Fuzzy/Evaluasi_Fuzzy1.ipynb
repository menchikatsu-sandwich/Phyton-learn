```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
usia_x=np.arange(0, 100, 0.5)

usia_rendah=fuzz.trapmf(usia_x,[0, 0, 5, 10])

fig, (ax0) = plt.subplots(nrows=1, figsize=(8,3))

ax0.plot (usia_x, usia_rendah, 'b', linewidth=1.5, label='Usia')

#mengisi judul grafik dengan usia
ax0.set_title('Usia')
#menampilkan legend keterangan graf
ax0.legend()
```

```python
#fungsi suhu demam ringan dengan trapizoid
suhu_x=np.arange(35, 43, 0.5)

suhu_demam_ringan=fuzz.trapmf(suhu_x,[37,37.5,38.5,39])

fig, (ax0) = plt.subplots(nrows=1, figsize=(8,3))

ax0.plot (suhu_x, suhu_demam_ringan, 'y', linewidth=1.5, label='Demam Ringan')

#mengisi judul grafik dengan suhu badan
ax0.set_title('Suhu Badan')
#menampilkan legend keterangan graf
ax0.legend()
```
