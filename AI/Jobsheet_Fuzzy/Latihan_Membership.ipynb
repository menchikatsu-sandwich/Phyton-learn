```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#rentang grafik sumbu x
suhu_x=np.arange(35, 43, 0.5)

#graf segitiga
suhu_normal=fuzz.trimf(suhu_x, [36.5, 37, 37.5])
#menyatakan medan gambar grafik
fig, (ax0) = plt.subplots(nrows=1, figsize=(8,3))
"""Menampilkan grafik ke layar, dimana sumbu x dari suhu x dan y dari perhitungan rumus fuzzy.trimf"""

ax0.plot (suhu_x,suhu_normal, 'g', linewidth=1.5, label='Normal')

#mengisi judul grafik dengan suhu badan
ax0.set_title('Suhu Badan')
#menampilkan legend keterangan graf
ax0.legend()
```

```python
#fungsi suhu rendah dengan triangle
suhu_x=np.arange(35, 43, 0.5)

suhu_rendah=fuzz.trimf(suhu_x,[0,35,36])

fig, (ax0) = plt.subplots(nrows=1, figsize=(8,3))

ax0.plot (suhu_x, suhu_rendah, 'b', linewidth=1.5, label='Suhu Rendah')

#mengisi judul grafik dengan suhu badan
ax0.set_title('Suhu Badan')
#menampilkan legend keterangan graf
ax0.legend()
```

```python
#fungsi suhu demam ringan dengan trapizoid
suhu_x=np.arange(35, 43, 0.5)

suhu_demam_ringan=fuzz.trapmf(suhu_x,[37,37.5,38.5,39])

fig, (ax0) = plt.subplots(nrows=1, figsize=(8,3))

ax0.plot (suhu_x, suhu_demam_ringan, 'y', linewidth=1.5, label='Demam Ringan')

#mengisi judul grafik dengan suhu badan
ax0.set_title('Suhu Badan')
#menampilkan legend keterangan graf
ax0.legend()
```

```python
#fungsi suhu demam tinggi dengan trapizoid
suhu_x=np.arange(35, 43, 0.5)

suhu_demam_tinggi=fuzz.trapmf(suhu_x,[38.5,40,42.5,42.5])

fig, (ax0) = plt.subplots(nrows=1, figsize=(8,3))

ax0.plot (suhu_x, suhu_demam_tinggi, 'r', linewidth=1.5, label='Demam Tinggi')

#mengisi judul grafik dengan suhu badan
ax0.set_title('Suhu Badan')
#menampilkan legend keterangan graf
ax0.legend()
```
