```python
# main program
# dict utk menyimpan penyakit
dict_penyakit = {
    "name": ["Flu", "Demam", "Bronkitis"],
}

# dict utk menyimpan gejala penyakit
dict_gejala = {
    "name": ["Panas", "Hidung Tersumbat"]
}

# dict utk menyimpan nilai gejala panas terhadap penyakit flu, demam, bronkitis
dict_nilai_M1 = {
    "name": [["F", "D", "B"], "Teta"],
    "value": [0.8, 0.2]
}

# dict utk menyimpan nilai gejala hidung tersumbat terhadap penyakit flu, demam, bronkitis
dict_nilai_M2 = {
    "name": [["A", "F", "D"], "Teta"],
    "value": [0.9, 0.1]
}

# dict utk menyimpan nilai penyakit berdasarkan gejala panas dan hidung tersumbat
result_nilai_M3 = {
    "name": [],
    "value": []
}

# program utama
for index in range(len(dict_nilai_M1["name"])):
    nameM1 = dict_nilai_M1["name"][index]
    valueM1 = dict_nilai_M1["value"][index]
    for index2 in range(len(dict_nilai_M2["name"])):
        nameM2 = dict_nilai_M2["name"][index2]
        valueM2 = dict_nilai_M2["value"][index2]

        #  mencari subset dari M1 dan M2
        if (nameM1 != "Teta") and (nameM2 != "Teta"):
            intersected_list = [element for element in nameM1 if element in nameM2]
            nameM3 = intersected_list
        elif (nameM1 == "Teta") and (nameM2 != "Teta"):
            nameM3 = nameM2
        elif (nameM1 != "Teta") and (nameM2 == "Teta"):
            nameM3 = nameM1
        elif (nameM1 == "Teta") and (nameM2 == "Teta"):
            nameM3 = "Teta"
        result_nilai_M3["name"].append(nameM3)

        # menghitung nilai M utk hasil subset
        valueM3 = valueM1 * valueM2
        result_nilai_M3["value"].append(valueM3)
print(result_nilai_M3)


# print nilai M3
for index in range(len(result_nilai_M3['name'])):
    print("Nilai kepercayaan penyakit ", result_nilai_M3['name'][index], " = ", result_nilai_M3['value'][index])

print("Nilai kepercayaan penyakit tertinggi adalah ", max(result_nilai_M3['value']))
```

```python
# main program
# dict utk menyimpan penyakit
dict_penyakit = {
    "name" : ["Flu", "Demam", "Bronkitis", "Alergi"],
}

# dict utk menyimpan gejala penyakit
dict_gejala = {
    "name" : ["Panas", "Hidung Tersumbat", "Piknik"]
}

# dict utk menyimpan nilai penyakit
# berdasarkan gejala panas dan hidung tersumbat
result_nilai_M3 = {
    "name" : [["F", "D"], ["F", "D", "B"], ["A", "F", "D"], "Teta"],
    "value" : [0.72, 0.08, 0.18, 0.02]
}

# dict utk menyimpan nilai gejala piknik
# terhadap penyakit alergi
dict_nilai_M4 = {
    "name" : ["A", "Teta"],
    "value" : [0.6, 0.4]
}

# dict temporary utk menyimpan nilai penyakit
# berdasarkan gejala panas, hidung tersumbat, dan piknik
temp_result_nilai_M5 = {
    "name" : [],
    "value" : []
}

# dict akhir utk menyimpan nilai penyakit
# berdasarkan gejala panas, hidung tersumbat, dan piknik
result_nilai_M5 = {
    "name" : [],
    "value" : []
}

# main program
for index in range(len(result_nilai_M3['name'])):
    nameM3 = result_nilai_M3["name"][index]
    valueM3 = result_nilai_M3["value"][index]
    for index in range(len(dict_nilai_M4['name'])):
        nameM4 = dict_nilai_M4["name"][index]
        valueM4 = dict_nilai_M4["value"][index]
        
        #mencari subset dari M3 dan M4
        if (nameM3 != "Teta") and (nameM4 != "Teta"):
            intersected_list = []
            for element in nameM3:
                if element in nameM4:
                    intersected_list.append(element)
            if intersected_list == []:
                nameM5 = "Null"
            else:
                nameM5 = intersected_list
        elif (nameM3 == "Teta") and (nameM4 == "Teta"):
            nameM5 = "Teta"
        elif (nameM3 == "Teta") and (nameM4 != "Teta"):
            nameM5 = nameM4
        elif (nameM3 != "Teta") and (nameM4 == "Teta"):
            nameM5 = nameM3
        
        #menambahkan hasil subset ke temp_result_nilai_M5
        temp_result_nilai_M5["name"].append(nameM5)
        #menghitung nilai M utk hasil subset
        valueM5 = valueM3 * valueM4
        #menambahkan hasil perhitungan ke M5
        temp_result_nilai_M5["value"].append(valueM5)

#code utk menggabungkan list string
for index in range(len(temp_result_nilai_M5["name"])):
    name = temp_result_nilai_M5["name"][index]
    if (isinstance(name, list) == True):
        name = ''.join(name)
    else:
        name = name
    temp_result_nilai_M5["name"][index] = name

#code utk mencari string unik dalam list_name
unique_values = list(set(temp_result_nilai_M5["name"]))

#code utk menjumlahkan seluruh nilai utk Null
sumNull = 0
for index in range(len(temp_result_nilai_M5['name'])):
    name = temp_result_nilai_M5["name"][index]
    if name == "Null":
        sumNull += temp_result_nilai_M5["value"][index]

#code utk menghitung nilai M5
for index in range(len(unique_values)):
    unique_name = unique_values[index]
    if (unique_name != "Null"):
        sumElement = 0
        for element in range(len(temp_result_nilai_M5["name"])):
            name = temp_result_nilai_M5["name"][element]
            if name == unique_name:
                sumElement += temp_result_nilai_M5["value"][element]
        value_M5 = sumElement / (1 - sumNull)
        if (unique_name != "Teta"):
            unique_name = list(unique_name)
        result_nilai_M5["name"].append(unique_name)
        result_nilai_M5["value"].append(value_M5)

#code utk mencetak nilai M5
for index in range(len(result_nilai_M5["name"])):
    print("Nilai kepercayaan penyakit ", result_nilai_M5["name"][index], " = ", result_nilai_M5["value"][index])

print("Nilai kepercayaan penyakit tertinggi adalah ", max(result_nilai_M5["value"]))
```
