```python
# func utk menghitung nilai CF 
# berdasarkan nilai MB dan MD
def processCF(MB, MD):
    return MB - MD

# func utk menghitung nilai MB 
# utk kombinasi evidence e1 dan e2
def processMBGabungan(MBhe1, MBhe2):
    return MBhe1 + MBhe2 * (1 - MBhe1)

# func utk menghitung nilai MD 
# utk kombinasi evidence e1 dan e2
def processMDGabungan(MDhe1, MDhe2):
    return MDhe1 + MDhe2 * (1 - MDhe1)

# main program
# dict utk menyimpan penyakit
dict_penyakit = {
    "name" : ["Cacar", "Alergi"],
}

# dict utk menyimpan gejala penyakit
dict_gejala = {
    "name" : ["Bintik"]
}

# dict utk menyimpan nilai MB, MD, dan CF 
# utk kombinasi evidence e1 dan e2
dict_penyakitGejala = {
    "name" : ["CacarBintik","AlergiBintik"],
    "MB": [0.8, 0.4],
    "MD": [0.01, 0.3],
    "CF" : []
}

# dict utk menyimpan nilai MB, MD, dan CF 
# suatu penyakit dengan kombinasi hipotesis h1 dan h2
result_Diagnose = {
    "name" : ["CacarAlergiBintik"],
    "MB_And" : [],
    "MD_And" : [],
    "CF_And" : [],
    "MB_Or" : [],
    "MD_Or" : [],
    "CF_Or" : []
}

# hitung CF utk suatu evidence e1
for index in range(len(dict_penyakitGejala['name'])):
    MB = dict_penyakitGejala["MB"][index]
    MD = dict_penyakitGejala["MD"][index]
    CF = processCF(MB, MD)
    dict_penyakitGejala["CF"].append(CF)

print(dict_penyakitGejala)

# hitung CF utk gabungan evidence e1 dan e2
for index in range(len(dict_penyakitGejala['name'])-1):
    MBh1 = dict_penyakitGejala["MB"][index]
    MDh1 = dict_penyakitGejala["MD"][index]
    MBh2 = dict_penyakitGejala["MB"][index+1]
    MDh2 = dict_penyakitGejala["MD"][index+1]

    MB_And = min(MBh1, MBh2)
    MD_And = min(MDh1, MDh2)
    CF_And = processCF(MB_And, MD_And)

    MB_Or = max(MBh1, MBh2)
    MD_Or = max(MDh1, MDh2)
    CF_Or = processCF(MB_Or, MD_Or)

    result_Diagnose["MB_And"].append(MB_And)
    result_Diagnose["MD_And"].append(MD_And)
    result_Diagnose["CF_And"].append(CF_And)

    result_Diagnose["MB_Or"].append(MB_Or)
    result_Diagnose["MD_Or"].append(MD_Or)
    result_Diagnose["CF_Or"].append(CF_Or)

print(result_Diagnose)

# print hasil ke dict
for index in range(len(result_Diagnose['name'])):
    print("CF Penyakit Cacar dan Alergi dengan Gejala Bintik = ", result_Diagnose["CF_And"][index])
    print("CF Penyakit Cacar atau Alergi dengan Gejala Bintik = ", result_Diagnose["CF_Or"][index])
```
