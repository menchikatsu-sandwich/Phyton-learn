```python
# hitung nilai real dari suatu bilangan biner
def to_real(binery,a,b,L):
    # konversi biner ke desimal (10)
    des = int(binery, 2)
    # hitung nilai real (0-255)
    x = a + ((b - a) / (2**L - 1)) * des
    return x
```

```python
# func calc fit kromosom
import math

def calculate_fitness(chromosome):
    # gabungkan setiap gen agar kromosom jadi biner
    binary = ''
    for i in range(len(chromosome)): # loop setiap gen
        binary += str(chromosome[i])
    # call func to_real
    x = to_real(binary,a,b,L)
    # hitung fx sesuai rumus
    return math.sin(22/7 * x / 256)
```

```python
import random

# funct generate random population
def generate_population(size, length):
    # inisialisasi populasi sbg list
    population = []
    # loop sebanyak size
    for _ in range(size):
        genes = [0,1] # isi gen 0 atau 1
        chromosome = []
        for _ in range(length): # ulang gen sebanyak gen (18)
            # isi sebuah kromosom dgn random 0 atau 1 sepanjang Legth
            chromosome.append(random.choice(genes))
        # buat individu dari kromosom dan fitness
        individu = {
            'chromosome': chromosome,
            'fitness': calculate_fitness(chromosome)
        }
        # tambah individu ke populasi
        population.append(individu)
        
    print("Generated a random pop of size:" , size)
    return population
```

```python
# funct untuk seleksi dgn teknik roulette wheel
# funct utk select 2 kromosom untuk crossover
import math

def select_chromosomes(population,crossover_probability):
    fitness_values = []
    for i in range(len(population)):
        fitness_values.append(population[i]['fitness'])
    # sum seluruh list dalam fitness_values
    probability = [float(fitness)/sum(fitness_values) for fitness in fitness_values]
    
    jumlah_kawin = math.ceil(crossover_probability * len(population))
    
    # ini akan memilih acak dari populasi sebanyak jumlah kawin
    # misal prob [0.5 0 0 0.3 0.2 0] maka item pertama akan terpilih 50% dari populasi
    parents = random.choices(population, weights=probability, k=jumlah_kawin)
    
    return parents
```

```python
# funct utk crossover 2 kromosom
def crossover(chromosome1, chromosome2):
    # membangkitkan bilangan rand dari 2 sampai sepanjang chromosom-3, yaitu 2 hngga 18-2 yaitu 16
    crossover_point = random.randint(2, len(chromosome1) - 2)
    # anak1 : induk satu ke-0 sampai crossover_point-1 + induk dua dari crossover_point ke akhir
    # anak2 : induk dua ke-0 sampai crossover_point-1 + induk satu dari crossover_point ke akhir
    child1_chromosome = chromosome1[0:crossover_point] + chromosome2[crossover_point:]
    child2_chromosome = chromosome2[0:crossover_point] + chromosome1[crossover_point:]
    
    # buat individu dari kromosom dan fitness
    child1 = {
        'chromosome': child1_chromosome,
        'fitness': calculate_fitness(child1_chromosome)
    }
    child2 = {
        'chromosome': child2_chromosome,
        'fitness': calculate_fitness(child2_chromosome)
    }
    
    return child1, child2
```

```python
# func utk mutasi kromosom
def mutate(chromosome):
    # membangkitkan  bilangan acak antara 0 hingga L-1 (17)
    for i in range(len(chromosome)):
        if random.uniform(0, 1) < mutation_probability: # jika random < mutation_probability maka gen kromosom diubah
            if chromosome[i] == 0:
                chromosome[i] = 1 # ubah 0 ke 1
            else:
                chromosome[i] = 0 # ubah 1 ke 0
    
    return chromosome
```

```python
# mengurutkan populasi descending berdasarkan fitness terbesar ke terkecil
from operator import itemgetter

def sort_population(population):
    population = sorted(population, key=itemgetter('fitness'), reverse=True)
    return population
```

```python
# func utk eliminasi
# menggunakan prinsip BGA untuk mempertahankan sejumlah individu dengan fitness terbaik
def eliminate(population, top_population, count):
    # count adalah jumlah individu yang bernilai tertinggi yang akan dipertahankan
    index = set()
    indexize = 0
    
    # mengacak sejumlah count bilangan yang harus dipastikan berbeda
    # misal count 4 maka membangkitkan 4 bilangan acak berbeda
    while indexize < count:
        # acak bil dari 0 hingga panjang populasi-1
        bil = random.randint(0, len(population) - 1)
        if bil not in index:
            indexize += 1
            index.add(bil)
            
    # gantikan sejumlah jumlah_lestari kromosom dengan top population
    j = 0
    for i in index:
        # gantikan individu i tertentu dgn individu terbaik (j)
        population[i] = top_population[j]
        j += 1
    return population
```

```python
# Tujuan adlh memaksimalkan nilai fx
# Hard constraint: nilai  x 0-255

# params
a = 0 # batas bawah
b = 255 # batas atas
L = 18 # panjang kromosom

# print funct pamars
print('a: {} b: {} L: {}'.format(a, b, L))

# params utk genetic algorithm
population_size = 100 # ukuran populasi
crossover_probability = 0.8 # probabilitas crossover
mutation_probability = 0.1 # probabilitas mutasi
generations = 50 # jumlah generasi
r_bga = 0.1 # rasio BGA
jumlah_lestari = int(population_size * r_bga) # jumlah individu yang dilestarikan

print('\nGenetic Algorithm Parameters:')
print('Population Size: ', population_size)
print('Mutation Probability: ', mutation_probability)
print('Crossover Probability: ', crossover_probability)
print('Generations: ', generations)
print('Value of r for BGA: {} sehingga jumlah individu yang dilestarikan {}'.format(r_bga, jumlah_lestari))

print('Performing genetic algorithm...\n')

# generate random population awal
population = generate_population(population_size, L)
# mengurutkan populasi berdasarkan fitness desc
population = sort_population(population)
# mengambil top population sesuai jumlah yg dilestarikan
top_population = population[0:(jumlah_lestari+1)]

population
```

```python
# melakukan pembangkitan setiap generasi
for i in range(generations):
    print('\n', 'Generations:', i+1)
    
    # menentukan jumlah perkawinan yaitu jumlah anak dikali 2 (bulatkan ke atas)
    # misal jumlah anak 5 maka jumlah kawin adalah ceil(5/2) = 3 perkawinan
    # pilih induk
    # pilih induk
    parents = select_chromosomes(population,crossover_probability)
    
    # hapus parents dari populasi
    for par in range(len(parents)):
        for popu in range(len(population)):
            if parents[par] == population[popu]:
                population.pop(popu) # keluarkan parent terpilih dari populasi
                break # keluar dari for kedua
    
    # kawin silang
    for cross in range(len(parents)//2):
        # mengawinkan parent 0 dengan 1, kemudian 2 dengan 3 dan seterusnya sampai habis
        child1, child2 = crossover(parents[0+2*cross]['chromosome'],parents[1+2*cross]['chromosome'])
        
        # mutasi
        child1['chromosome'] = mutate(child1['chromosome'])
        child2['chromosome'] = mutate(child2['chromosome'])
        
        # menghitung fitness anak setelah dimutasi
        child1['fitness'] = calculate_fitness(child1['chromosome'])
        child2['fitness'] = calculate_fitness(child2['chromosome'])
        
        # menghitung fitness anak setelah dimutasi anak ke dalam generasi
        population.append(child1)
        population.append(child2)
    
    # BGA
    # melaksanakan proses BGA (menggantikan beberapa individu acak dengan list individu terbaik)
    population = eliminate(population, top_population, jumlah_lestari)
    # mengurutkan kembali populasi secara descending berdasarkan fitness
    population = sort_population(population)
    # pastikan ukuran populasi tetap 100
    population = population[0:100]
    # menyimpan individu terbaik sejumlah testari
    top_population = population[0:(jumlah_lestari+1)]
    
    # mencetak satu individu terbaik pada setiap generasi
    print('Best Individu:', population[0])

population
```
