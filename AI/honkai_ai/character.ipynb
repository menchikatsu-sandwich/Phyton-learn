```python
# UJI COBA EKSTRAK DATA PERTAMA
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# Path ke ChromeDriver (sesuaikan dengan lokasi di komputer kamu)
service = Service("C:/Users/santo/Documents/chromedriver-win64/chromedriver.exe")
driver = webdriver.Chrome(service=service)

# URL halaman karakter
url = "https://honkai-star-rail.fandom.com/wiki/Category:Playable_Characters"
driver.get(url)

# Tunggu sampai elemen kategori karakter muncul (gunakan XPath atau class yang benar)
try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "category-page__member"))
    )
    print("Halaman berhasil dimuat!")
except:
    print("Timeout! Halaman tidak dimuat dengan benar.")
    driver.quit()
    exit()

# Ambil HTML setelah render
soup = BeautifulSoup(driver.page_source, "html.parser")

# Cari semua elemen karakter (cari elemen yang sesuai dari HTML)
characters = soup.select(".category-page__member .category-page__member-link")

print("=== DATA KARAKTER ===")
for char in characters:
    name = char.text.strip()  # Ambil nama karakter dari <a>
    link = "https://honkai-star-rail.fandom.com" + char["href"]  # Ambil link ke karakter
    print(f"Nama: {name}, Link: {link}")

# Tutup browser setelah selesai
driver.quit()
```

```python
# UJI COBA EKSTRAK DATA DETAIL KARAKTER
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import csv
import time

# Path ke ChromeDriver (sesuaikan dengan lokasi di komputer kamu)
service = Service("C:/Users/santo/Documents/chromedriver-win64/chromedriver.exe")
driver = webdriver.Chrome(service=service)

# URL halaman karakter
url = "https://honkai-star-rail.fandom.com/wiki/Category:Playable_Characters"
driver.get(url)

# Tunggu sampai elemen kategori karakter muncul
try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "category-page__member"))
    )
    print("Halaman berhasil dimuat!")
except:
    print("Timeout! Halaman tidak dimuat dengan benar.")
    driver.quit()
    exit()

# Ambil HTML setelah render
soup = BeautifulSoup(driver.page_source, "html.parser")

# Cari semua elemen karakter
characters = soup.select(".category-page__member .category-page__member-link")

# List untuk menyimpan data karakter
character_data = []

# Loop melalui setiap karakter dan ambil detailnya
for char in characters:
    name = char.text.strip()  # Ambil nama karakter dari <a>
    link = "https://honkai-star-rail.fandom.com" + char["href"]  # Ambil link ke karakter
    
    # Kunjungi halaman karakter untuk mengambil detail
    driver.get(link)
    time.sleep(2)  # Tunggu sebentar agar halaman dimuat
    
    char_soup = BeautifulSoup(driver.page_source, "html.parser")
    
    # Ambil data dari bagian Bio
    combat_path = char_soup.select_one(".pi-data-value[data-source='path'] a")
    combat_type = char_soup.select_one(".pi-data-value[data-source='combat'] a")
    species = char_soup.select_one(".pi-data-value[data-source='species']")
    factions = char_soup.select_one(".pi-data-value[data-source='factions']")
    world = char_soup.select_one(".pi-data-value[data-source='worlds']")
    release_date = char_soup.select_one(".pi-data-value[data-source='releasedate']")
    
    # Pastikan data yang diambil tidak kosong
    combat_path = combat_path.text.strip() if combat_path else "N/A"
    combat_type = combat_type.text.strip() if combat_type else "N/A"
    species = species.text.strip() if species else "N/A"
    factions = factions.text.strip() if factions else "N/A"
    world = world.text.strip() if world else "N/A"
    release_date = release_date.text.strip() if release_date else "N/A"
    
    # Tambahkan data karakter ke list
    character_data.append([name, combat_path, combat_type, species, factions, world, release_date])

# Tutup browser setelah selesai
driver.quit()

# Simpan data karakter ke file CSV
with open("character_data.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Nama", "Combat Path", "Combat Type", "Species", "Factions", "World", "Release Date"])
    writer.writerows(character_data)

print("Data karakter berhasil disimpan ke character_data.csv")
```
