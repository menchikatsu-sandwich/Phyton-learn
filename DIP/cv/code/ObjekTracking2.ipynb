```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Tidak dapat membuka kamera.")
    exit()

# Ambil frame pertama dan konversi ke grayscale
ret, frame1 = cap.read()
if not ret:
    print("Gagal membaca frame pertama.")
    cap.release()
    exit()

prvs = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)

hsv = np.zeros_like(frame1)
hsv[..., 1] = 255  # Saturasi maksimal untuk visualisasi warna

while True:
    ret, frame2 = cap.read()
    if not ret:
        print("Gagal membaca frame berikutnya.")
        break

    next = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

    # Hitung optical flow Farneback
    flow = cv2.calcOpticalFlowFarneback(prvs, next, None,
                                        pyr_scale=0.5, levels=3, winsize=15,
                                        iterations=3, poly_n=5, poly_sigma=1.2, flags=0)

    # Konversi flow ke polar untuk visualisasi
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1], angleInDegrees=True)
    
    hsv[..., 0] = ang/2
    hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)  # Value sesuai magnitude

    rgb = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

    cv2.imshow('Farneback Optical Flow', rgb)

    # Tekan ESC untuk keluar
    if cv2.waitKey(1) & 0xFF == 27:
        break

    prvs = next

cap.release()
cv2.destroyAllWindows()
```
