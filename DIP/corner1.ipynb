```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread('mobil.jpg')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_flat_chess, cmap='gray')
```

```python
# konversi nilai integer
# dalam setiap pixel mejadi float32
gray = np.float32(gray_flat_chess)
# dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
# nilai valiable kSize defaultnya adalah 3
# karena merupakan hasil perhitungan atas r score, sedangkan nilai k=0.04 ini bisa diubah, yang akan memeengaruhi sens corner detect
dst = cv2.cornerHarris(gray, 2, 3, 0.1)

dst = cv2.dilate(dst, None)
flat_chess[dst > 0.01 * dst.max()] = [255, 0, 0]
plt.imshow(flat_chess)
```
