```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

flat_chess = cv2.imread('mobil.jpg')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)

gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_flat_chess, cmap='gray')

# Syntax:
# cv2.goodFeaturesToTrack(image, maxCorners, qualityLevel, minDistance)
# Parameters:
# image - The source image we need to extract the features from
# maxCorners - Maximum number of corners we want 
#              (0 value gives all the corners)
# qualityLevel - Quality level parameter (preferred value = 0.01)
# minDistance - Minimum distance (preferred value = 10)

# Shi-Tomasi => goodFeaturesToTrack

corners = cv2.goodFeaturesToTrack(gray_flat_chess, 0, 0.01, 10)
corners = np.intp(corners)

for corner in corners:
    x, y = corner.ravel()
    cv2.circle(flat_chess, (x, y), 3, 255, -1)

plt.imshow(flat_chess)
```
