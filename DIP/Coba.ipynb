```python
import cv2
import matplotlib.pyplot as plt

# Membaca gambar
img = cv2.imread('mobil.jpg')

# OpenCV membaca gambar dalam format BGR, kita ubah ke RGB agar warna benar
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Menampilkan gambar di Jupyter Notebook
plt.imshow(img_rgb)
plt.axis('off')  # Hilangkan axis agar lebih bersih
plt.show()
```

```python
import cv2
import numpy as np

# Baca gambar
image = cv2.imread("asszilla.png")  # Ganti dengan path gambar yang ingin diuji

# Definisi kernel (contoh: edge detection)
custom_kernel = np.array([
    [0, 1, 0],
    [1, 0, 1],
    [0, 1, 0]
], dtype=np.float32)

# Terapkan filter ke gambar
filtered_image = cv2.filter2D(image, -1, custom_kernel)

# Tampilkan hasil
cv2.imshow("AssZilla", image)
cv2.imshow("AssNigga", filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
