```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('orghytam.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

hist = cv2.calcHist([img_rgb], [0], None, [256], [0, 256])
plt.plot(hist, color='r')
plt.xlim([0, 256])
plt.show()
```

```python
#Static
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
std_dev = np.std(gray)

print(f"Standar Deviasi: {std_dev}")
if std_dev > 50:
    print("Gambar memiliki kontras tinggi.")
else:
    print("Gambar memiliki kontras rendah.")
```

```python
#dynamic
min_val = np.min(gray)
max_val = np.max(gray)
range_val = max_val - min_val

print(f"Rentang Intensitas: {range_val}")
if range_val > 100:
    print("Kontras tinggi.")
else:
    print("Kontras rendah.")
```
