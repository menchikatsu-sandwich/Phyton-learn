```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cleansed_sales_1500.csv")
summary = df.groupby('item_type')['units_sold'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
summary.plot(kind='bar', color='mediumseagreen')
plt.title("Total Units Sold per Item Type (1500 Records)")
plt.xlabel("Item Type")
plt.ylabel("Units Sold")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("units_sold_1500.png")
plt.show()
```
