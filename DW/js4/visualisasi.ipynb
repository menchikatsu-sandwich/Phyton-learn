```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('datamentah.csv')

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Quantity'], df['TotalAmount'], color='blue', alpha=0.5)
plt.title('Hubungan antara Quantity dan TotalAmount')
plt.xlabel('Quantity')
plt.ylabel('TotalAmount')
plt.grid(True)
plt.show()

# Histogram
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['Quantity'], bins=10, color='green', alpha=0.7)
plt.title('Distribusi Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frekuensi')
plt.subplot(1, 2, 2)
plt.hist(df['TotalAmount'], bins=10, color='orange', alpha=0.7)
plt.title('Distribusi TotalAmount')
plt.xlabel('TotalAmount')
plt.ylabel('Frekuensi')
plt.tight_layout()
plt.show()

# Bar chart
sales_by_product = df.groupby('ProductID')['TotalAmount'].sum()
plt.figure(figsize=(10, 6))
sales_by_product.plot(kind='bar', color='purple', alpha=0.7)
plt.title('Total Penjualan per Product')
plt.xlabel('ProductID')
plt.ylabel('TotalAmount')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
